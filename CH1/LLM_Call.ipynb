```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from the .env file
load_dotenv()

# The client automatically looks for the "OPENAI_API_KEY" environment variable
client = OpenAI()

try:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": "Explain pdf files in one sentence."}
        ]
    )
    print(response.choices[0].message.content)

except Exception as e:
    print(f"Error: {e}")
```

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

## Here using a free model from google

# 1. Explicitly load the variables from your .env file
load_dotenv()

# 2. (Optional) Run a quick sanity print statement to check it loaded
# print("Loaded Key prefix:", os.environ.get("GEMINI_API_KEY")[:6])

# 3. Initialize your LangChain architecture
gpt = init_chat_model(model="gemini-2.5-flash", model_provider="google_genai")

response = gpt.invoke("Can you list out some tamil comedies")
print(response.content)
```
