```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} {role}"),
    ("user", "Write a fun fact about {topic}. Make sure to have the key value pair. Key is 'fact' and value is the response"),
])

user_input = input("Enter any topic for getting the fun fact")
user_tone = input("Enter the tone:")
user_role = input("Enter the role for me:")

ready_prompt = prompt_template.invoke({"topic":user_input, "tone":user_tone, "role":user_role})
ready_prompt
```

```python
gpt = init_chat_model(model="gemini-2.5-flash", model_provider="google_genai")

gpt.invoke(ready_prompt).content
```

### *Guiding in Prompts*

```python
response = gpt.invoke("Tell me a joke. Generate the output in key value pair with the following keys: setup, punchline. Also format it with no extra symbols")
response.content
```

### ***Structured outputs using pydantic models***

```python
from pydantic import BaseModel, Field

##Basemodel is a class that we use to create our schema

class llm_schema(BaseModel):
    setup:str = Field(description="setup for the joke")
    punchline:str = Field(description="punchline for the joke")

```

```python
obj = llm_schema(**{"setup": "some setup", "punchline": "some punchline"})
obj
print(obj)
```

```python
llm_structured_output = gpt.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me a joke")

result
```
