### Prompts are the dynamic way to send messages. And they are user friendly 

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
import os
```

```python
if os.environ.get("GEMINI_API_KEY"):
    load_dotenv()
    print("GEMINI API Key is available")
else:
    raise ValueError
    print("No API key")
```

```python
user_input = input("For what do you need a sql query?")

dynamic_prompt = PromptTemplate.from_template("Write a sql query for finding {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

```

```python
gpt = init_chat_model(model="gemini-2.5-flash", model_provider="google_genai")

gpt.invoke(ready_prompt).content
```
