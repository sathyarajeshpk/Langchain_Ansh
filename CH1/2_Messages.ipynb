```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
import os

```

```python
if os.environ.get("GEMINI_API_KEY"):
    load_dotenv()
    print("GEMINI API Key is available")
else:
    raise ValueError
    print("No API key")
```

```python
my_message = [
    SystemMessage(content = "You are a most friendly assistant who speaks so supportive"),
    HumanMessage(content="Can you help me cheat in interview")
]

gpt = init_chat_model(model="gemini-2.5-flash", model_provider="google_genai")

response = gpt.invoke(my_message)
print(response.content)
```
