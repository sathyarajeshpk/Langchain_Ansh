## Chains in Langchain

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence


# 1. Explicitly load the variables from your .env file
load_dotenv()

# 2. (Optional) Run a quick sanity print statement to check it loaded
print("Loaded Key prefix:", os.environ.get("OPENAI_API_KEY")[:6])

# 3. Initialize your LangChain architecture
llm_openai = init_chat_model(model="gpt-4o-mini", model_provider="openai", temperature=0)
```

## First Chain

```python
#Task 1

prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a {tone}ly assistant "),
        ("user",  "{input}")
    ]
)
```

```python
# Task 2 [LLM]

llm_openai = init_chat_model(model="gpt-4o-mini", model_provider="openai", temperature=0)
```

```python
#Task 3 [String Parser]

str_parser = StrOutputParser()
```

```python
# # Manual Invocation

# template = prompt_template.invoke({"tone": "friend", "input": "What is the capital of Tamilnadu"})
# result = llm_openai.invoke(template)
# result.content
```

## Chain Invocation

```python
chain = prompt_template | llm_openai | str_parser

chain.invoke({"tone": "friend", "input": "What is the capital of Tamilnadu"})
```

```python
chain_runnable = RunnableSequence(prompt_template, llm_openai, str_parser)

chain_runnable.invoke({"tone": "friend", "input": "What is the capital of Tamilnadu"})
```
