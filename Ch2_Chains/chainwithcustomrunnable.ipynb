## Custom runnable

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence, RunnableLambda


# 1. Explicitly load the variables from your .env file
load_dotenv()

# 2. (Optional) Run a quick sanity print statement to check it loaded
print("Loaded Key prefix:", os.environ.get("OPENAI_API_KEY")[:6])

# 3. Initialize your LangChain architecture
llm_openai = init_chat_model(model="gpt-4o-mini", model_provider="openai", temperature=0)
```

```python
#Task 1

prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a {tone}ly assistant "),
        ("user",  "{input}")
    ]
)
```

```python
# Task 2 [LLM]

llm_openai = init_chat_model(model="gpt-4o-mini", model_provider="openai", temperature=0)
```

```python
#Task 3 [String Parser]

str_parser = StrOutputParser()
```

```python
# Task 4 - Custom Runnable

def dictionary_maker(text:str)->dict:

    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

```python
#Task 5 - Again a prompt using the output of custom runnable

prompt_template2 = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a online content creator. Make sure to exclude the \n kind of values from the output. Give output in a single paragraph"),
        ("user",  "Create a post for the following: {text}")
    ]
)
```

```python
#Task 6 - Post Generation

llm_openai = init_chat_model(model="gpt-4o-mini", model_provider="openai", temperature=0)
```

```python
# Task 7 - Str parser to user

str_parser = StrOutputParser()
```

```python
chain = prompt_template | llm_openai | str_parser | dictionary_maker_runnable | prompt_template2 | llm_openai | str_parser

chain.invoke({"tone": "friend", "input": "What is the capital of Tamilnadu"})
```
